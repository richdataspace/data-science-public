# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/1mhXZ3JPAnekXP_4vX7Z_sJj35VWqayaR?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time
import pyperclip
import pandas as pd
```

```python
# Passo 1: Acessar no sistema do google drive
pyautogui.PAUSE = 1

#abre uma aba nova
pyautogui.hotkey('command', 't')

#acessa a url
pyautogui.hotkey('command', 'l')

#define link
link = 'https://drive.google.com/drive/folders/1wRTFw0sUVBjRr4hW5U9LF7DjLixRyxym'

#copia e cola
pyperclip.copy(link)
pyautogui.hotkey('command', 'v')

#acessa com enter
pyautogui.press('enter')

#esperar um pouquinho
time.sleep(5)

# Passo 2: Entrar na pasta da Aula 1
pyautogui.click(894, 355, button='secondary')

# Clica no 3 pontinhos
pyautogui.click(1232, 189)

# Passo 3: Fazer download da base

# Download
pyautogui.click(1080, 588)

# Aguardar o download
time.sleep(5)
```

```python
# Passo 4: Calcular Faturamente e Qt de pedidos
tabela = pd.read_excel(r'/Users/guilhermedelima/Downloads/Vendas - Dez.xlsx', engine='openpyxl')
```

```python
tabela.head()
```

```python
faturamento = tabela['Valor Final'].sum()
```

```python
qtde_produtos = tabela['Quantidade'].sum()
```

```python
# Passo 5: Entrar no meu email 
email = 'guilherme.barbosa.ufpr@gmail.com'

#abre uma aba nova
pyautogui.hotkey('command', 't')

time.sleep(0.5)

#acessa a url
pyautogui.hotkey('command', 'l')

#define link
linkgmail = 'https://mail.google.com'

#copia e cola
pyperclip.copy(linkgmail)
pyautogui.hotkey('command', 'v')

#acessa com enter
pyautogui.press('enter')

time.sleep(2)

#clica no escrever
pyautogui.click(75, 202)
time.sleep(5)

#preenche o email
pyautogui.write(email)

time.sleep(1)

pyautogui.press('tab')
time.sleep(1)

#vai pro proximo campo do Assunto
pyautogui.press('tab')
time.sleep(1)

pyautogui.write(r"Esse eh o assunto do email - Relatorio de Vendas")

# vai pro corpo do email
pyautogui.press('tab')

# Passo 6: Criar o email

texto_email = (f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f},

A quantidade de pedidos foi: {qtde_produtos:,}

Abracos, Guilhermino
""")

pyperclip.copy(texto_email)
pyautogui.hotkey('command', 'v')

#esperar um pouquinho
time.sleep(2)


# Passo 7: Enviar email
# pyautogui.click(813, 869)
# pyautogui.hotkey('command', 'enter')
```
